```python
%matplotlib qt
import numpy as np
import pandas as pd
from pandas import DataFrame
from scipy import stats
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
import gc
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier,RandomForestRegressor
from sklearn.externals import joblib
from sklearn.metrics import confusion_matrix
import warnings
warnings.filterwarnings('ignore')
```

```python
contemporary_y = 2018

if 'data' in locals():
    print('data is prepared……')
else :
    data = pd.read_excel(r'../../小需求/fillna/y/2013-2018_y.xlsx')
    weight = joblib.load(r'../../小需求/fillna/y/weight_dict_y.m')
    grid = joblib.load('../../小需求/fillna/y/剔除华信永泰.m')[0]
```

```python
# col = ['名称','报告期','获息倍数','货币资金/短期债务','净利润(亿元)','筹资活动现金流(亿元)','主营业务利润率(%)',
#        '主营业务收入增长率(%)','投资活动现金流(亿元)','净资产回报率(%)','总资产报酬率(%)','EBITDA/带息债务','流动比率',]
# col = ['名称','报告期','获息倍数','货币资金/短期债务','净利润(亿元)','筹资活动现金流(亿元)','主营业务利润率(%)',
#        '主营业务收入增长率(%)','经营活动现金流(亿元)','净资产回报率(%)','总资产报酬率(%)','EBITDA/带息债务','流动比率',]
col = ['名称','报告期','获息倍数','货币资金/短期债务','净利润(亿元)','筹资活动现金流(亿元)','主营业务利润率(%)','投资活动现金流(亿元)',
       '主营业务收入增长率(%)','经营活动现金流(亿元)','净资产回报率(%)','总资产报酬率(%)','EBITDA/带息债务','流动比率',]
data_contemporary = data.loc[data['报告期'].dt.year == (contemporary_y-1), col].copy()

re = pd.read_excel('../../data/Debt issuing company 2018 report/report of defaulted.xlsx')
re = re.groupby(['名称'], as_index=False).apply(lambda x: x.sort_values(by=['发生日期'], ascending = True).iloc[0, :])

re_contemporary =  re.loc[re['发生日期'].dt.year == contemporary_y, :]

re_past = re.loc[re['发生日期'].dt.year < contemporary_y, '名称'].tolist()
re_out = re_contemporary.loc[~re_contemporary['名称'].isin(data_contemporary['名称']), '名称'].tolist()
re_drop = re_past + re_out

data_contemporary = data_contemporary.loc[~data_contemporary['名称'].isin(re_drop), :]
data_contemporary_re = data_contemporary.loc[data_contemporary['名称'].isin(re_contemporary['名称']), :]
data_contemporary_re.insert(0, '0-1 score', grid.predict_proba(data_contemporary_re.iloc[:, 2:])[:, 1])
accu_ = sum(data_contemporary_re['0-1 score'] >= 0.5)/data_contemporary_re.shape[0]

y_true = data_contemporary['名称'].isin(re_contemporary['名称'])
print(f'year:{contemporary_y},accu:{accu_},\
      \nconfusion_matrix:\n{confusion_matrix(y_true,grid.predict(data_contemporary.iloc[:,2:]))}')
```

```python
def rechange(dataframe, col, weight):
    if col in weight.keys() :
        if weight[col] == 'log1p':
            return np.exp(dataframe[col]) - 1
        else:
            return dataframe[col] * weight[col][1] + weight[col][0]
    else :
        return dataframe[col]

for i in data_contemporary_re.columns:
    data_contemporary_re[i] = rechange(data_contemporary_re,i,weight)
    
data_contemporary_re
```

```python
all_data = data_contemporary.copy()
all_data.insert(0, '0-1 score', grid.predict_proba(data_contemporary.iloc[:,2:])[: ,1])
for i in all_data.columns:
    all_data[i] = rechange(all_data,i,weight)
all_data.loc[all_data['0-1 score'] >= 0.5, :].to_excel('./123.xlsx',index=False)
```
