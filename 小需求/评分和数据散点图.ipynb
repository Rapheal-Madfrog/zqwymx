```python
%matplotlib qt
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_lr = pd.read_excel('/lr.xlsx')
data_rf = pd.read_excel('/rf.xlsx')
```

```python
list_ = ['货币资金/短期债务','净利润(亿元)',
                 '主营业务利润率(%)','主营业务收入增长率(%)',
                 '净资产回报率(%)','总资产报酬率(%)','流动比率',]
```

```python
lr_coef = [-0.015771  , -0.45104409, -0.06339978, -0.41327249, -1.77518834,
        -0.37779527,  0.13244211]
lr_coef = np.array(lr_coef)[:,np.newaxis]
data_lr['all'] = data_lr.iloc[:,1:-1].dot(lr_coef).values
```

```python
rf_fea_importance = [0.26063124, 0.1904248 , 0.0901712 , 0.12993525, 0.18160121,
       0.09717059, 0.0500657 ]
rf_fea_importance = np.array(rf_fea_importance)[:,np.newaxis]
data_rf['all'] = data_rf.iloc[:,1:-1].dot(rf_fea_importance).values
```

```python
plt.figure()
for i in range(data_lr.shape[1]-1):
    ax = plt.subplot(4,2,i+1)
    ax.scatter(x=data_lr.iloc[:,i],y=data_lr.iloc[:,0],)
    ax.legend(loc='best',prop={'size':11})
#     ax.set_xlim([-1,1])
#     ax.set_ylim([0,1])
    ax.set_title(list(data_lr)[i+1])
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')
    ax.set_xlabel('')
```

```python
plt.figure()
for i in range(data_rf.shape[1]-1):
    ax = plt.subplot(4,2,i+1)
    ax.scatter(x=data_rf.iloc[:,i],y=data_rf.iloc[:,0],)
    ax.legend(loc='best',prop={'size':11})
#     ax.set_xlim([-1,1])
#     ax.set_ylim([0,1])
    ax.set_title(list(data_rf)[i+1])
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')
    ax.set_xlabel('')
```
