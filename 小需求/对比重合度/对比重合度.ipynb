```python
import pandas as pd
from functools import reduce
```

```python
re270 = pd.read_excel(path+r'\nian\评分的270家.xlsx','re270')
re270 = re270[[i for i in re270.columns if i.find('Unnamed')==-1]]
re270 = re270.sort_values(by=list(re270)[0],ascending=False)
re270.index = (range(re270.shape[0]))
```
